```python
def extractColumn(data, column):
    columnData = []

    columnIndex = -1

    if(column in data[0]):
        columnIndex = data[0].index(column)
    else:
        return None

    for index, line in enumerate(data):
        if(index != 0):
            for index, col in enumerate(line):
                if(index == columnIndex):
                    columnData.append(col)

    return columnData
```

# 9

- Loop through each row in legislators.
  - Parse the birth year from the birthday column.
  - Convert the birth year to an integer, and assign it to birth_year.
  - Wrap this code in a try/except block.
    - If there's an exception, assign 0 to birth_year.
  - Append birth_year to the row with the append() method.
- When finished, legislators should have an extra column for birth year.


```python
import csv

legislators = list(csv.reader(open("legislators.csv")))

birth_year = []

for index, row in enumerate(legislators):
    if(index != 0):
        birthYear = 0
        birthComponents = row[2].split('-')

        try:
            birthYear = int(birthComponents[0])
        except Exception as e:
            birthYear = 0

        birth_year.append(birthYear)
        row.append(birthYear)
```
