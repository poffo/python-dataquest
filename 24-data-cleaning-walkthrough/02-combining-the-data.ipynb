```python
import pandas as pd
import numpy as py
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
```

```python
data_files = [
        "ap_2010.csv",
        "class_size.csv",
        "demographics.csv",
        "graduation.csv",
        "hs_directory.csv",
        "sat_results.csv"
]
```

```python
data = {}

for file in data_files:
    fileName = file.split(".")
    temp_data = pd.read_csv("schools/" + file)
    data[fileName[0]] = temp_data
```

```python
all_survey = pd.read_csv("schools/survey_all.txt", delimiter="\t", encoding="windows-1252")

data["survey"] = all_survey
```

```python
def paddingZeros(v):
    v = v.zfill(2);
    return v

data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(str)
data["class_size"]["padded_csd"] = data["class_size"]["padded_csd"].apply(paddingZeros)
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]

class_size = data["class_size"].loc[(data["class_size"]["GRADE "] == "09-12") & (data["class_size"]["PROGRAM TYPE"] == 'GEN ED')]
```

```python
class_size = class_size.groupby("DBN").aggregate(py.mean)

class_size.head()
```

```python
class_size.reset_index(inplace=True)
```

```python
data["class_size"] = class_size
print data["class_size"].head(5)
```

```python
#14

# Calculate the means of all of the columns in combined using the pandas.DataFrame.mean() method.
# Fill in any missing values in combined with the means of the respective columns using the pandas.DataFrame.fillna() method.
# Fill in any remaining missing values in combined with 0 using the df.fillna() method.
# Display the first few rows of combined to verify that the correct operations occurred.


for column in combined.columns:
    if(is_numeric_dtype(combined[column])):
        combined[column].fillna(combined[column].mean(), inplace=True)
        combined[column].fillna(0, inplace=True)

# mean = combined.mean()
# combined.fillna(mean, inplace=True)
# combined.fillna(0, inplace=True)
```
